## こういうデータベースを作成してみる
| id | name | price |
| - | - | - |
|  |  |  |
|  |  |  |
|  |  |  |

```python
import sqlite3
path = ''
db_name = 'car.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = 'CREATE TABLE cars (id INT, name TEXT, price Real);'

    cur.execute(sql)
except sqlite3.Error as e:
    print('error', e )
finally:
    conn.close()
```

```python
path = ''
db_name = 'car.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = "INSERT INTO cars (id, name, price) VALUES (1, '370z', 7000000);"

    cur.execute(sql)

    conn.commit() 
except sqlite3.Error as e:
    print('error', e)
finally:
    conn.close()
```

```python
path = ''
db_name = 'car.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = "INSERT INTO cars (id, name, price) VALUES (?, ?, ?);"
    cars = [
        (2, '911', 15500000),
        (3, 'GR Supra', 8000000),
        (4, 'Type-R', 5000000)
    ]

    cur.executemany(sql, cars)

    conn.commit() 
except sqlite3.Error as e:
    print('error', e)
finally:
    conn.close()
```

```python
path = ''
db_name = 'car.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = "SELECT * FROM cars;"

    cur.execute(sql)
except sqlite3.Error as e:
    print('error', e)
else:
    for row in cur:
        id, name, price = row
        print(id, name , price)

finally:
    conn.close()
```

```python
path = ''
db_name = 'car.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = "DELETE FROM cars WHERE id = 1;"

    cur.execute(sql)
    conn.commit()
except sqlite3.Error as e:
    print('error', e)

finally:
    conn.close()
```

```python
path = ''
db_name = 'car.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = "UPDATE cars SET price = ? WHERE id = ?"

    cur.execute(sql, (5000000, 2))

    conn.commit()
except sqlite3.Error as e:
    print('error', e)

finally:
    conn.close()
```
